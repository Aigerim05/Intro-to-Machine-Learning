## Step 13 : What are advantages and disadvantages of such encoding of a categorical varable? Does the answer depend on whether it is nominal or ordinal?

```python

# 1. OHE's advantages:  allows algorithms like linear models to learn relationships between features 
# more easily, as there’s no risk of misleading the model into interpreting a non-existent ordinal relationship.

# OHE's disadvantage: If the categorical variable has many unique categories, 
# One-Hot Encoding can lead to high dimensionality. 

# 2. OHE works good on nominal data, because it treats the categories as independent. So, OHE will ignore
#  the order for ordinal data
```

## Step 14 : What can be said about linear dependence of the columns produced by one-hot-encoding? Consider two cases: with and without dropping.

```python
# Without dropping a category: The columns produced by One-Hot Encoding are linearly dependent. 
# This is because the information from one column can be inferred from the others, leading to multicollinearity.

# With dropping a category: The columns are linearly independent. By dropping one category, we remove this redundancy, 
# and the remaining columns provide unique information about the categories relative to the baseline category.
```

## Step 16 : Normalize some feature to a range [0, 1]

```python
import pandas as pd

df_bike = pd.read_csv("Bike.csv").rename(columns={'cnt' : 'cnt_rental_bike'})

```

```python
# using sklearn

from sklearn import preprocessing 

x = df_bike[['cnt_rental_bike']]
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)

df_bike['cnt_rental_bike'] = x_scaled
df_bike['cnt_rental_bike'].describe()
```

```python
# using pandas
column = df_bike['cnt_rental_bike']
normalized_df = (column- column.min())/(column.max() - column.min())

normalized_df.describe()
```

## Step 17 : Standardize the same Feature(s) using standardization

```python
import pandas as pd

df_bike = pd.read_csv("Bike.csv").rename(columns={'cnt' : 'cnt_rental_bike'})

```

```python
# using pandas
column = df_bike['cnt_rental_bike']
standardized_df = (column - column.mean())/column.std()

standardized_df.describe()
```

```python
# using sklearn

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# StandardScaler expects a 2D array
df_bike['cnt_rental_bike'] = scaler.fit_transform(df_bike['cnt_rental_bike'].values.reshape(-1, 1))

df_bike['cnt_rental_bike'].describe()
```
